Zamysł: Jako synonimy będziemy brać pod uwagę wyrazy w nawiasach, dodatkowo ujęte w ''' '''.

```python
file = open('poczatki_wikipediowe.txt')
lines_list = file.readlines()

clean_lines_list = []
for line in lines_list:
    clean_line = line.replace('\n','')
    clean_lines_list.append(clean_line)
```

```python
dic_def = {}
l = len(clean_lines_list)

# słownik - hasło, definicja
for i in range(0,l,3):
    dic_def[clean_lines_list[i].replace('### ', '')] = clean_lines_list[i+1]

# słownik - hasło, synonimy
dic_syn = {}
for i in range(0,l,3):
    dic_syn[clean_lines_list[i].replace('### ', '')] = []
```

```python
hasla = [clean_lines_list[i].replace('### ', '') for i in range(0,l,3)]
definicje = [clean_lines_list[i] for i in range(1,l,3)]
synonimy = ['' for i in range(len(hasla))]
```

```python
for i in range(len(definicje)):
    d = definicje[i]
    if '(' in d:
        synonimy[i] = d[d.find("(") + 1:d.find(")")].split()
    else:
        synonimy[i] = []
```

```python
#for i in range(len(synonimy)):
#   j = 0
#   while j < len(synonimy[i]):
#        if ("'''" in synonimy[i][j]):
#            synonimy[i][j:j+1] = [' '.join(synonimy[i][j:j+1]).replace("'''",'')]
#        j += 1
```

```python
#lista = [[] for i in range(len(hasla))]

for i in range(len(synonimy)):
    for s in synonimy[i]:
        if ("'''" in s):
            s = s.replace("'''",'')
            if s is not hasla[i]: # bierzemy pod uwagę tylko wyrażenia w ''''''
                #lista[i].append(s)
                dic_syn[hasla[i]].append(s)
```

```python
import unicodedata
import numpy as np
```

```python
plik = open('znaki_wikipedii.txt')
lines = plik.readlines()
#for x in lines:
 #   print(x)
```

```python
clean_lines_list = []

for line in lines:
    clean_line = line.replace('\n',' ')
    clean_line = clean_line.replace('  ', '')
    clean_lines_list += clean_line.split(' ')
    
#for x in clean_lines_list:
 #   print(unicodedata.category(x))
```

##### Zbiory o tym samym typie:

```python
dictt = {}
for x in clean_lines_list:
    cat = unicodedata.category(x)
    if cat in dictt: # unikać .keys()
        dictt[cat].append(x)
    else:
        dictt[cat] = [x]
```

```python
for k, v in dictt.items():
    print(k, v)
```

##### Dodatkowe dzielenie klastrów zawierających litery:

```python
#for v in dictt.values():
  #  for x in v:
   #     print(x, unicodedata.bidirectional(x), unicodedata.name(x))
    #print('\n')

letters = []
letters_keys = []

for k, v in dictt.items():
    if 'L' in k:
        letters.append(v)
        letters_keys.append(k)
        #print([ord(x) for x in v])
```

```python
ord_letters = [[] for i in range(len(letters))]

for i in range(5):
    for l in letters[i]:
        ord_letters[i].append(ord(l))
    ord_letters[i].sort()
```

```python
letters_keys
```

```python
min_max = [(min(ord_letters[i]), max(ord_letters[i])) for i in range(len(ord_letters))]
min_max
```

```python
kernels = []
for x in min_max:
    kernels.append(np.arange(0, x[1], 255//2+1))
    
for i in range(len(kernels)):
    kernels[i] = kernels[i][1::2]
    
kernels[3]
```

```python
def nearest_kernel(given_value, kernel_list):
    absolute_difference_function = lambda list_value : abs(list_value - given_value)
    closest_value = min(kernel_list, key = absolute_difference_function)
    return closest_value
```

```python
near_kernels = [[],[],[],[],[]]

for i in range(5):
    for j in range(len(ord_letters[i])):
        near_kernels[i].append(nearest_kernel(ord_letters[i][j], kernels[i]))
```

```python
file = open('cytaty.txt')
lines = file.readlines()
#len(lines)
```

```python
for i in range(4,20):
    print(lines[i])
```

```python
from nltk import word_tokenize

def my_tokenizer(text):
    text = text.replace('\n', '')
    #text = text.replace('&lt;br /&gt;', ' ')
    text_list = text.split()
    punctuations = '''.,;:'...'?!()[]„”'"-'''
    for word in text_list:
        text_list[text_list.index(word)] = word.strip(punctuations)
    for word in text_list:
        if word == '':
            text_list.remove(word)
    return text_list

#lines[8], my_tokenizer(lines[8]), word_tokenize(lines[8])
for i in range(4,20):
    print(my_tokenizer(lines[i]), '\n', word_tokenize(lines[i]))
    print('\n')
```

Funkcja word_tokenize oddziela znaki interpunkcyjne znajdujące się na początku lub końcu wyrazu, funkcja my_tokenizer je usuwa.

#### Lepszy tokenizator: 
- oprócz tego co jest w funkcji my_tokenizer, dzieli tekst w miejscu w którym jest coś innego niż str lub liczba, np: flaga biało-czerwona -> flaga biało czerwona
- zamiana &lt;br /&gt; na spację
